```python
#12. Write a Python program to create multiple types of charts
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([10, 20, 30, 40, 50])
y2 = np.array([50, 40, 30, 20, 10])
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Chart
axs[0, 0].plot(x, y1, marker='o', linestyle='-', color='b', label="Line 1")
axs[0, 0].plot(x, y2, marker='s', linestyle='--', color='r', label="Line 2")
axs[0, 0].set_title("Line Chart")
axs[0, 0].legend()

# Bar Chart
axs[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'purple', 'orange'])
axs[0, 1].set_title("Bar Chart")

# Scatter Plot
axs[1, 0].scatter(x, y1, color='g', label="Data 1")
axs[1, 0].scatter(x, y2, color='m', label="Data 2")
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].legend()

# Pie Chart
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple', 'orange'])
axs[1, 1].set_title("Pie Chart")

# Adjust layout and show
plt.tight_layout()
plt.show()

```

```python
#10
import matplotlib.pyplot as plt
import numpy as np

# Sample data
groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
men_means = [22, 30, 35, 35, 26]
women_means = [25, 32, 30, 35, 29]

# X positions for each group
x = np.arange(len(groups))

# Width of bars
width = 0.35

# Create bar chart for men and women
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, men_means, width, label='Men', color='b')
bars2 = ax.bar(x + width/2, women_means, width, label='Women', color='r')

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Groups')
ax.set_ylabel('Scores')
ax.set_title('Scores by Group and Gender')
ax.set_xticks(x)
ax.set_xticklabels(groups)
ax.legend()

# Display the plot
plt.show()


```

```python
#2
import matplotlib.pyplot as plt
import numpy as np

mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_velocity = [0.3314, 0.2278, 0.2836, 0.2645]

x = np.arange(len(mean_velocity))


plt.bar(x, mean_velocity, yerr=std_velocity, capsize=5, color='skyblue', edgecolor='blue')


plt.title('Bar Plot with Error Bars')
plt.xlabel('Measurement Points')
plt.ylabel('Mean Velocity')


plt.xticks(x, ['Point 1', 'Point 2', 'Point 3', 'Point 4'])


plt.show()
```
